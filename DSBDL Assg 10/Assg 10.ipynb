```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df = pd.read_csv("iris.csv")
```

```python
print("\nðŸ“Œ First 5 Rows of the Dataset:")
print(iris_df.head())
```

```python
print("\nðŸ“Œ Features and Their Data Types:")
print(iris_df.dtypes)
```

```python
plt.figure(figsize=(12, 6))
iris_df.drop(columns=["species"]).hist(figsize=(12, 6), bins=20, color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions - Histograms", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(iris_df.columns[:-1]):  # Exclude species
    plt.subplot(2, 2, i+1)
    sns.boxplot(y=iris_df[col], color="lightblue")
    plt.title(f"Boxplot of {col}")
plt.suptitle("Feature Spread - Boxplots", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
print("\nðŸ“Œ Identified Outliers Using IQR:")
for col in iris_df.columns[:-1]:  # Ignore 'species' as it's categorical
    Q1 = iris_df[col].quantile(0.25)
    Q3 = iris_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = iris_df[(iris_df[col] < lower_bound) | (iris_df[col] > upper_bound)]
    print(f"ðŸ”¹ {col}: {len(outliers)} outliers found")
```
