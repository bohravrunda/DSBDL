```python
 import seaborn as sns
 import matplotlib.pyplot as plt
 import pandas as pd
 # Load the Titanic dataset
 titanic = sns.load_dataset('titanic')
 titanic.head()
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
 import pandas as pd
 # Step 1: Fill missing 'deck' values with mode (most common deck)
 titanic['deck'] = titanic['deck'].fillna(titanic['deck'].mode()[0])
 # Step 2: Drop rows where 'embarked' or 'embark_town' is missing
 titanic_cleaned = titanic.dropna(subset=['embarked', 'embark_town'])
 # Step 3: Impute missing 'age' values with the median age
 median_age = titanic_cleaned['age'].median()
 titanic_cleaned.loc[:, 'age'] = titanic_cleaned['age'].fillna(median_age)
 # Step 4: Verify that there are no more missing values
 missing_values = titanic_cleaned.isnull().sum()
 print("Missing Values After Cleaning:\n", missing_values)
```

```python
titanic_cleaned.describe()
```

```python
 plt.figure(figsize=(6,4))
 sns.countplot(x='survived', hue='survived', data=titanic_cleaned, 
palette='viridis')  
# viridis Lower values = dark colors,Higher values = bright colors
 plt.title('Count of Survivors vs. Non-Survivors')
 plt.xlabel('Survived (0 = No, 1 = Yes)')
 plt.ylabel('Count')
 plt.show()
```

```python
plt.figure(figsize=(8,6 ))
sns.countplot(x='pclass', hue='survived' , data=titanic_cleaned, palette='viridis')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel( 'Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic_cleaned, x='age', bins=30, color='teal', kde=True)

plt.title('Age Distribution of Passengers with KDE')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(
    data=titanic_cleaned,
    x='age',
    hue='survived',
    bins=30,
    palette='viridis',
    kde=True,
    multiple='dodge',
    edgecolor='black',
    alpha=0.7  # Transparency level
)

plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')

# Customize the legend
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set figure size

# Histogram for 'fare' distribution
sns.histplot(
    titanic_cleaned['fare'],
    kde=True,  # Add KDE
    bins=40,   # Set number of bins
    color='purple'  # Set color to purple
)

# Title and axis labels
plt.title('Fare Distribution')
plt.xlabel('Fare')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Set figure size

# Countplot for survival rate based on embarkation port
sns.countplot(
    data=titanic_cleaned,
    x='embarked',  # X-axis: embarkation port
    hue='survived',  # Hue: survival status
    palette='viridis'  # Color palette
)

# Title and axis labels
plt.title('Survival Count by Embarkation Port')
plt.xlabel('Embarked (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Count')

# Customize the legend
plt.legend(title='Survived', loc='upper right')

# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")  # Set the style for the plot

# Defining the number of bins for the histogram
num_bins = 30  

# Create the histogram plot
plt.figure(figsize=(12, 6))
sns.histplot(
    data=titanic_cleaned,
    x='fare',
    hue='survived',
    bins=num_bins,
    kde=True,             # Adding a KDE overlay
    palette='pastel',    
    edgecolor='black',    
    alpha=0.7             # Transparency level
)

# Adding title and labels
plt.title('Distribution of Ticket Prices by Survival Status', fontsize=16)
plt.xlabel('Fare ($)', fontsize=14)
plt.ylabel('Number of Passengers', fontsize=14)

# Customizing legend
plt.legend(title='Survived', labels=['No', 'Yes'])

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```
